```python
import numpy as np
import pandas as pd

df = pd.read_csv('paska.txt',sep='\t',header=None, names=["text"])
df
```

```python
df["text"] = df["text"].str.replace("\d+", "")
df['text'] = df['text'].str.replace('[^a-zA-Z0-9\s]', '').str.lower()
df
```

```python
stopwords = ["a", "an", "and", "as"," at", "for", "from", "in", "into", "of", "on", "or", "the", "to"]

df['text'] = df['text'].str.split().apply(lambda x: ' '.join([w for w in x if w not in stopwords]))
df["text"]
```

```python
df['text'] = df['text'].str.split().apply(lambda x: ' '.join([w for w in x if len(w) >= 2]))
df['text']
```

```python
df['words'] = df['text'].str.split()
df['words']
```

```python
from nltk.util import pad_sequence
from nltk.util import bigrams
from nltk.util import ngrams
from nltk.util import everygrams
from nltk.lm.preprocessing import pad_both_ends
from nltk.lm.preprocessing import flatten

for row in df['words']:
    print(list(bigrams(row)))
```

```python
from nltk.util import pad_sequence
padded_sent = list(pad_sequence(df['words'][0], pad_left=True, left_pad_symbol="<s>", 
                                pad_right=True, right_pad_symbol="</s>", n=2))
list(ngrams(padded_sent, n=2))
```
